## Pivoting and Grouping Data with Pandas

### Pivot Tables

- Useful for summarizing data
- Create pivot tables with `pivot_table()` method
- Key parameters are:
    - `data`: the dataframe to be pivoted
    - `values`: columns to be aggregated
    - `index`: column to be used as rows (summarized by)
    - `columns`: columns to be used as columns in the pivot table
    - `aggfunc`: aggregation function to be applied to the `values`
    - `fill_value`: for null values

### Grouping

- allows you to group rows based on one or more columns e.g. average number of deaths by age group
- Key parameter: `by`: column by which you want to group by
- Aggregation functions: `sum()`, `mean()`, `median()`
