## Film Data

````python
# Select the title, rating and rental duration from the film table
film_query = """SELECT *
                FROM film
            """

film_df = pd.read_sql(film_query, db_cnx)
film_df.head()
````

````python
# Select everything in the film catergories table
film_categories = """SELECT *
                    FROM film_category
                """

film_categories_df = pd.read_sql(film_categories, db_cnx)
film_categories_df.head()
````

````python
# PYTHON ONLY: Join the film data and film category and return as a dataframe 
joined_film_py = film_df.merge(film_categories_df, on = "film_id")
joined_film_py.head(2)
````

````python
# SQL ONLY: Join the film data and film category and return as a dataframe 
film_join_query = """
                    SELECT *
                    FROM film as fm
                    JOIN film_category as fc
                    ON fm.film_id = fc.film_id
                """

joined_film_df = pd.read_sql(film_join_query, db_cnx)
joined_film_df.head()
````

````python
# WARNING!: This is a (complex?) join, and could get confusing
# It begins with a join of the film category table to the film table.
# Then the film table is joined to the inventory table
# The inventory table is then joined to the rental table
# All the above are then joined to the customer data
# Essentially, for each customer, we will have data on their interactions with films and rentals
joined_data_query = """SELECT * 
                        FROM film_category as fc
                        JOIN film as fm
                        ON fc.film_id = fm.film_id
                        JOIN inventory as iv
                        ON fm.film_id = iv.film_id
                        JOIN rental as rn
                        ON iv.inventory_id = rn.inventory_id
                        JOIN customer as cu
                        ON rn.customer_id = cu.customer_id
                        """

crazy_joined_df = pd.read_sql(joined_data_query, db_cnx)
crazy_joined_df
````

## Slicing Timeseries

When working with timeseries data always set the datetime as index of the dataframe

- Set the date column of interest as index
- Ensure that it is sorted in ascending order
- Perform your slicing, with the period set as a string. Use the syntax "YYYY/MM/DD HH:MM"

````python
# Lets plot the payment series
august_data= cust_pay_df.set_index("payment_date").sort_index(ascending=True).loc["2005/08/01 00:00":"2005/08/31 23:59"].interpolate()
august_data["amount"].plot()
plt.title("Trend of Transactions in August 2005")
````

> - Because the index is a time series, we can leverage the ability to **resample.** 

> - In time series, **resampling** simply means changing the frequency of observations. It could either be upward (months to days) or downward (days to months)

````python
# Total Daily Payments
cust_pay_df.set_index("payment_date").sort_index(ascending=True)["amount"].resample("D").sum()
````

````python
# Total Monthly Payments
cust_pay_df.set_index("payment_date").sort_index(ascending=True)["amount"].resample("M").sum().loc["2006/01":"2006/02"].plot(alpha=0.2)
````

````python
# Total Daily Payments May - June
cust_pay_df.set_index("payment_date").sort_index(ascending=True)["amount"].resample("D").sum().loc["2005/5":"2005/6"].plot(figsize=(10, 4), alpha=.5, linewidth=3, label="Daily Totals - May to June")
````

**Re: Key Parameters in the plot above:**
- figsize = (l,w): used to set a size of the image you are plotting
- alpha = `decimal`: used to set how light or dark the figures (bar or line) in the plot appears
- linewidth = `int`: for a line plot, it is used to determine how thick or thin the lines are
- label = `str`: used to set the title of the plot

> - A common operation in time series is to plot rolling 7-day average data on top of daily data. The .rolling method accepts
a required argument `window` (the size of the moving window).

> It then allows you to aggregate the result, like the groupby function

````python
# 7-day Rolling total payments
pd.Series(cust_pay_df.set_index("payment_date").sort_index(ascending=True)["amount"].resample("D").sum().rolling(7).median()).head(10)
````

````python
# 7-day Rolling total payments
pd.Series(cust_pay_df.set_index("payment_date").sort_index(ascending=True)["amount"].resample("D").sum().rolling(5).median()).head(10)
````

````python
# Rolling Payments
cust_pay_df.set_index("payment_date").sort_index(ascending=True)["amount"].resample("M").sum().loc["2005/5":"2005/7"].plot.barh()
````
